### Typical 1. Total Time to Watch a Series

Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.

**Example Output:**
```
Total time: 39h 30m 0s
```


````python
# Your code here

season1_minutes = 45
season1_seconds = 35
season1_total_seconds = (season1_minutes * 60 + season1_seconds) * 10

season2_5_minutes = 35
season2_5_seconds = 15
season2_5_total_seconds = (season2_5_minutes * 60 + season2_5_seconds) * 10 * 4

total_seconds = season1_total_seconds + season2_5_total_seconds

hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

# Output result
print(f"Total time: {hours}h {minutes}m {seconds}s")
````

### Excellent 2. Data Science: Mini Statistics Report

Ask for five numeric values (space-separated). Print the values, total, mean, median, and sample standard deviation.

**Example Output:**
```
Values: 80, 90, 70, 85, 75
Total = 400.00
Mean = 80.00
Median = 80.00
Std Dev = 7.91
```


*Optional hint:* `import statistics as stats`


````python
# Your code here

import statistics as stats

# Ask for 5 numeric values (space-separated)
input_values = input("Enter 5 numbers (space-separated): ")

# Convert to list of floats
numbers = list(map(float, input_values.split()))

# Check if exactly 5 numbers were entered
if len(numbers) != 5:
    print("Error: Please enter exactly 5 numbers.")
else:
    total = sum(numbers)
    mean = stats.mean(numbers)
    median = stats.median(numbers)
    std_dev = stats.stdev(numbers)  # sample standard deviation

    # Print results
    print("Values:", ", ".join(f"{n:.0f}" for n in numbers))
    print(f"Total = {total:.2f}")
    print(f"Mean = {mean:.2f}")
    print(f"Median = {median:.2f}")
    print(f"Std Dev = {std_dev:.2f}")
````

