4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
p_to_kg = 0.45359237
in_to_m = 0.0254

w = float(input("Weight (lb): "))
h = float(input("Height (in): "))

w_kg = w * p_to_kg
h_m = h * in_to_m

bmi = w_kg / (h_m ** 2)

print(f"\nBMI: {bmi:.2f}")

if bmi < 18.5:
    print("Underweight")
elif bmi <= 25.0:
    print("Normal")
elif bmi <= 30.0:
    print("Overweight")
else:
    print("Obese")
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
n = int(input("Enter number of students: "))
marks = [0] * n

for i in range(n):
    marks[i] = float(input(f"Enter mark for student {i+1}: "))

total = 0
for m in marks:
    total += m
avg = total / n

high = max(marks)
low = min(marks)

pass_count = 0
for m in marks:
    if m >= 50:
        pass_count += 1
fail_count = n - pass_count

print(f"\nAverage: {avg:.2f}")
print(f"Highest: {high}")
print(f"Lowest: {low}")
print(f"Passed: {pass_count}")
print(f"Failed: {fail_count}")
````

12. **Dog Years**

It is commonly said that one human year is equivalent to 7 dog years. However, this simple conversion fails to recognise that dogs reach adulthood in approximately two years. As result, some people believe that it is better to count each of the first two human years as 10.5 dog years, and then count each additional human year as 4 dog years. 

Write a program that implements the conversion from human years to dog years described in the previous paragraph. Ensure that your program works correctly for conversions of less than two or more humans’ years. Your program should display an appropriate error message if the user enters a negative number. 


````python
human_years = float(input("Enter human years: "))

if human_years < 0:
    print("Error: Age cannot be negative.")
else:
    if human_years <= 2:
        dog_years = human_years * 10.5
    else:
        dog_years = 2 * 10.5 + (human_years - 2) * 4

    print(f"{human_years} human years is equivalent to {dog_years} dog years.")
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
import random

secret = random.randint(1, 100)
tries = 0

print("Guess the secret number between 1 and 100!")

while True:
    guess = int(input("Enter your guess: "))
    tries += 1

    if guess < secret:
        print("Too small!")
    elif guess > secret:
        print("Too large!")
    else:
        print(f"Congratulations! You guessed it in {tries} tries.")
        break
````

18. Suppose you want to develop a **program to play a lottery.**
    
The program randomly generates a two-digit number, prompts the user to enter a <u>two-digit number</u>, and determines whether the user wins according to the following rules:


    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.



````python
import random

lottery = random.randint(10, 99)
print("Lottery number is generated.")

user = input("Enter a two-digit number: ")

if len(user) != 2 or not user.isdigit():
    print("Please enter exactly two digits.")
else:
    print("Lottery number:", lottery)

    if user == str(lottery):
        print("Exact match! You win £10,000!")
    elif user[0] in str(lottery) and user[1] in str(lottery):
        print("All digits match! You win £3,000!")
    elif user[0] in str(lottery) or user[1] in str(lottery):
        print("One digit matches! You win £1,000!")
    else:
        print("No match, better luck next time!")
````


19. The program prompts the user to enter a <u>three-digit number</u> and determines whether the user wins according to the following rules:

    1. If the user input matches the lottery number in the exact order, the award is £10,000.
    2. If all the digits in the user input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user input matches a digit in the lottery number, the award is £1,000.

````python
import random

lottery = random.randint(100, 999)
print("Lottery number is generated.")

user = input("Enter a three-digit number: ")

if len(user) != 3 or not user.isdigit():
    print("Please enter exactly three digits.")
else:
    print("Lottery number:", lottery)
    lottery_str = str(lottery)

    if user == lottery_str:
        print("Exact match! You win £10,000!")
    elif sorted(user) == sorted(lottery_str):
        print("All digits match! You win £3,000!")
    elif any(digit in lottery_str for digit in user):
        print("One digit matches! You win £1,000!")
    else:
        print("No match, better luck next time!")
````

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
#Ali Hassan
#MISIS(M01096927)

n = int(input("Enter number of students: "))

highest = -1
second_highest = -1

for _ in range(n):
    score = float(input("Enter student score: "))
    
    if score > highest:     # Update highest and second highest when a new top score is found
        second_highest = highest
        highest = score
    elif score > second_highest and score != highest:     # Update only the second highest if it's between the two
        second_highest = score

print(f"Highest score: {highest}")
if second_highest == -1: # Handle case when there is no valid second highest score
    print("There is no second highest score.")
else:
    print(f"Second highest score: {second_highest}")
````
